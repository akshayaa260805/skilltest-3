```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Step 1: Create dataset (Voltage, Current)
data = np.array([
    [4.9, 0.88],
    [5.3, 0.92],
    [4.6, 1.05],
    [4.8, 0.85],
    [12.4, 2.30],
    [11.8, 2.10],
    [12.2, 2.35],
    [11.5, 1.95],
    [12.1, 2.25]
])

# Step 2: Apply K-Means
model = KMeans(n_clusters=2, random_state=42)
model.fit(data)

# Step 3: Extract cluster results
cluster_ids = model.predict(data)
centers = model.cluster_centers_

print("Cluster Assignment:", cluster_ids)
print("Cluster Centers:\n", centers)

# Step 4: Visualization
plt.figure()

# Plot each cluster separately
for i in range(2):
    plt.scatter(data[cluster_ids == i, 0],
                data[cluster_ids == i, 1],
                label=f'Cluster {i}')

# Plot centroids
plt.scatter(centers[:, 0], centers[:, 1],
            marker='X', s=200, label='Centroids')

plt.xlabel("Voltage")
plt.ylabel("Current")
plt.title("K-Means Clustering Result")
plt.legend()
plt.grid()

plt.show()
```
