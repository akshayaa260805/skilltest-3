```python
import numpy as np
import matplotlib.pyplot as graph
from sklearn.linear_model import LinearRegression

# Input data (years and corresponding house prices)
years = np.array([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018])
prices = np.array([25, 28, 30, 34, 37, 40, 43, 45, 47])

# Reshape years for model training
years_reshaped = years.reshape(-1, 1)

# Create and train the regression model
lr_model = LinearRegression()
lr_model.fit(years_reshaped, prices)

# Estimate price for a future year
future = np.array([[2019]])
estimated_value = lr_model.predict(future)

print("Estimated House Price in 2019:", round(estimated_value[0], 2), "Lakhs")

# Visual representation
graph.scatter(years, prices, label="Actual Data")
graph.plot(years, lr_model.predict(years_reshaped), linestyle="--", label="Trend Line")

graph.xlabel("Year")
graph.ylabel("Price (Lakhs)")
graph.title("House Price Growth Over Years")
graph.legend()
graph.show()
```
