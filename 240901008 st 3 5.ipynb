```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Create a different dataset
data = {
    'Device': ['D1', 'D2', 'D3', 'D1', 'D2', 'D3', 'D1', 'D2', 'D3', 'D1'],
    'Power': [60, 45, 85, 70, 55, 95, 65, 50, 90, 75],
    'Voltage': [220, 180, 240, 210, 190, 250, 230, 200, 260, 225],
    'Status': ['Stable', 'Unstable', 'Stable', 'Stable', 'Unstable',
               'Stable', 'Stable', 'Unstable', 'Stable', 'Stable']
}

df = pd.DataFrame(data)

print("Dataset:\n", df)

# Encode categorical data
device_map = {'D1': 0, 'D2': 1, 'D3': 2}
df['Device'] = df['Device'].map(device_map)

status_map = {'Stable': 1, 'Unstable': 0}
df['Status'] = df['Status'].map(status_map)

# Features and target
features = ['Device', 'Power', 'Voltage']
X = df[features]
y = df['Status']

# Create and train model
model = DecisionTreeClassifier(max_depth=3)
model.fit(X, y)

# Plot decision tree
plt.figure(figsize=(10,6))
tree.plot_tree(model, feature_names=features, class_names=['Unstable', 'Stable'], filled=True)
plt.title("Decision Tree for Device Stability")
plt.show()
```
